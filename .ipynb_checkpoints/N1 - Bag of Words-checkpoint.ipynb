```python
import pandas as pd
```

```python
messages = pd.read_csv('dataset/spam.csv', names=['label', 'message'])
```

```python
messages
```

```python
## Data Cleaning and Pre-processing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
# Task - implement the above with snowball stemmer
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
```

```python
snow_stemmer = SnowballStemmer('english') 
corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-z0-9]', ' ',messages['message'][i])
    review = review.lower()
    review = [snow_stemmer.stem(word) for word in review.split() if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
# Create a bag of words model
from sklearn.feature_extraction.text import CountVectorizer
#for Binary BOW, we gotta enable binary=True
# max_features refers to the number of most frequent words
cv = CountVectorizer(max_features = 500, binary=True)

X = cv.fit_transform(corpus).toarray()
```

```python
## Pre-processing with Lemmatization
## Then BOW
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []

for i in range(len(messages)):
    review = re.sub('[^a-zA-z0-9]', ' ', messages['message'][i])
    review = review.lower()
    review = [lemmatizer.lemmatize(word) for word in review.split() if word not in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
# create bag of words model
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 500, binary=True)

X = cv.fit_transform(corpus).toarray()
```

```python
# create bag of words model
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 300, binary=True, ngram_range=(2,2))
# ngram_range(2,2) -> only consider bigrams
# ngram_range(2,3) -> consider bigrams and trigrams
# ngram_range(1,3) -> consider unigrams, bigrams and trigrams
X = cv.fit_transform(corpus).toarray()
```
